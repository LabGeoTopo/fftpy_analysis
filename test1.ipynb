```octave

%read obs
filename='A_BIG.csv'
[num,txt,raw]=dlmread(filename);
a=csv2cell(filename);
datestr(a(:,10),dd/mm/yyyy);

%fid = fopen(filename)
%data = textscan(fid,'%s%f%f%f%f%f%f%f%f%s','Delimiter',',','HeaderLines',1);
%fclose(fid)
%t=cell2table(data)
t=0:0.001:2;
```

```octave
x=2*sin(20*pi*t) + sin(100*pi*t);
subplot(2,1,1)
plot(1000*t,x)
grid
xlabel("Time in milliseconds")
ylabel("Signal amplitude")
```

```octave
subplot(2,1,2)
y=fft(x);
plot(1000*t,abs(y))
xlabel("Frequency")
ylabel("Signal amplitude")
```

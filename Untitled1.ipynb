```python
import math

## GRS-80 constants, see 
## https://geodesy.geology.ohio-state.edu/course/refpapers/00740128.pdf
f_grs80 = .00335281068118
a_grs80 = 6378137e0

def geodetic2geocentric_torge(lat, a=a_grs80, f=f_grs80):
  ## note that [1 - eccentricity^2]^(1/2) = 1 - f
  return math.atan((1e0-f)*(1e0-f)*math.tan(lat))

def geodetic2geocentric_matlab(lat, hell=0e0, a=a_grs80, f=f_grs80):
  print(lat)
  ecc2 = f / (2e0-f)
  print(ecc2)
  N = a /( math.sqrt(1e0 - ecc2 * math.sin(lat) * math.sin(lat)))
  print(N)
  rho = (N+hell) * math.cos(lat)
  print(rho)
  z = (N*(1-ecc2) + hell) * math.sin(lat)
  print(z)
  return math.atan(z/rho)

angle = .1 ## degrees
print('{:10s} {:10s} {:10s}'.format('Geodetic', 'Torge Alg.', 'Matlab Alg.'))
while angle < 90e0:
  print('{:10.3f} {:10.3f} {:10.3f}'.format(angle, 
    math.degrees(geodetic2geocentric_torge(math.radians(angle))), 
    math.degrees(geodetic2geocentric_matlab(math.radians(angle)))))
  angle += 5 ## step = 5 degrees
```
